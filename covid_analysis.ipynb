```python
import pandas as pd
```

```python
data=pd.read_excel('Downloads//owid-covid-data.xlsx')
```

```python
data.head()
```

```python
summed=data.groupby('location').sum()
df=summed.reset_index()
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
To_plot=['Nepal','India','China','Bhutan']
```

```python
import matplotlib.pyplot as plt
```

```python
subdata=df[df['location'].isin(['Nepal','India','China','Bhutan'])]
```

```python
subdata[['location','total_cases','new_cases','total_deaths','new_deaths']]
```

```python
#fig ,axarr=plt.subplots(2,2,figsize=(12,8))
```

```python
subdata.plot(kind='line',x='location',y='new_cases',fontsize=16)
plt.title('Confirmed Cases')
plt.ylabel('new_cases')
plt.xlabel('country')
plt.legend()
plt.show()
subdata.plot(kind='bar',x='location',y='new_deaths',color='red',fontsize=16)
plt.title('new_death analysis')
plt.xlabel('location')
plt.ylabel('new_deaths')
plt.show()
subdata.plot(kind='scatter',x='location',y='total_cases',fontsize=16)
plt.xlabel('location')
plt.ylabel('total_cases')
plt.title('total cases analysis')
plt.show()
subdata.plot(kind='hist',x='location',y='total_deaths')
plt.xlabel('location')
plt.ylabel('total_deaths')
plt.title('total death analysis')
plt.show()

```
